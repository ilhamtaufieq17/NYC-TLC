<b>2b. Filling Missing Value</b>

<b>1. Handle invalid passenger counts</b> 

```python
df['passenger_count'] = df['passenger_count'].replace(0, 1)
df['passenger_count'] = df['passenger_count'].apply(lambda x: min(x, 6))
df['passenger_count'].fillna(df['passenger_count'].mode()[0], inplace=True)
```

In this section, the code:

- Replaces passenger counts of 0 with 1 (assuming a minimum of 1 passenger per trip)
- Caps passenger counts at 6 (This information is based on [This](https://www.nyc.gov/site/tlc/passengers/passenger-frequently-asked-questions.page) page)
- Fills missing passenger counts with the most frequent value (mode) in the column

<b>2. Adjust fare amounts</b>

```python
df['fare_amount'] = df['fare_amount'].apply(lambda x: 3 if x < 3 else x)
```

Here, the code sets a minimum fare amount of $3. If the fare amount is less than $3, it is adjusted to $3.

<b>3. Handle Other Missing Value</b>

```python
df['RatecodeID'] = df['RatecodeID'].fillna(value=99)
df['payment_type'].fillna(df['payment_type'].mode()[0], inplace=True)
df['trip_type'].fillna(df['trip_type'].mode()[0], inplace=True)
```

- Fills missing `RatecodeID` values with a default value of 99
- Fills missing `payment_type` and `trip_type` values with their respective modes (most frequent values)

<b>4. Update fare amounts based on RatecodeID</b>

```python
df.loc[df['RatecodeID'] == 2, 'fare_amount'] = 70
```

Here, the code sets the fare amount to $70 for trips with a `RatecodeID` of 2. `RatecodeID` = 2 means the trip is start or end in JFK.

<b>5. Update congestion surcharge based on pickup/dropoff locations</b>

```python
df.loc[(df['Borough_pick'] == 'Manhattan') | (df['Borough_drop'] == 'Manhattan'), 'congestion_surcharge'] = 2.75
df['congestion_surcharge'].fillna(0, inplace=True)
```

Here, the code updates the location information (borough, zone, and service zone) for specific location IDs (264 and 265) to either "Unknown" or "Outside of NYC". Other than that, fill the NaN value with 0. This change refer to [this](https://home.nyc.gov/site/tlc/passengers/taxi-fare.page) article 

<b>6. Update location information for specific location IDs</b>

```python

df.loc[df['PULocationID'] == 264, ['Borough_pick', 'Zone_pick', 'service_zone_pick']] = 'Unknown'
df.loc[df['DOLocationID'] == 264, ['Borough_drop', 'Zone_drop', 'service_zone_drop']] = 'Unknown'
df.loc[df['PULocationID'] == 265, ['Borough_pick', 'Zone_pick', 'service_zone_pick']] = 'Outside of NYC'
df.loc[df['DOLocationID'] == 265, ['Borough_drop', 'Zone_drop', 'service_zone_drop']] = 'Outside of NYC'
```

Here, the code updates the location information (borough, zone, and service zone) for specific location IDs (264 and 265) to either "Unknown" or "Outside of NYC".

```python
avg_trip_distance = round(df[df['trip_distance']>0]['trip_distance'].mean(), 3)
avg_fare_amount = round(df[df['fare_amount']>0]['fare_amount'].mean(), 3)

avg_price_per_mile = round(avg_fare_amount/avg_trip_distance, 2)
avg_price_per_mile
```

This code calculates the average price per mile for a taxi trip dataset stored in a Pandas DataFrame df. It:

Calculates the average trip distance and fare amount, filtering out zero values.
Divides the average fare amount by the average trip distance to get the average price per mile.
Rounds the result to 2 decimal places.
The final result is stored in the variable avg_price_per_mile.

```python
distance = df['fare_amount']/avg_price_per_mile
```

this code calculates a new value called `distance` based on two existing values: `fare_amount` and `avg_price_per_mile`.

```python
df['trip_distance'] = np.where(df['trip_distance']==0, distance, df['trip_distance'])
```

This code updates the `trip_distance` column in the DataFrame df by replacing any values that are equal to 0 with the value stored in the distance variable. If the value is not 0, it remains unchanged.

<b>1. When are the peak and off-peak times for taxi demand throughout the day and week?</b>

```python
hourly_trips = df.groupby('pickup_hour').size()
daily_trips = df.groupby('pickup_day_of_week').size()

days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_trips = daily_trips.reindex(days_order)

plt.figure(figsize=(14, 6))
plt.plot(hourly_trips.index, hourly_trips.values, marker='o', linestyle='-', color='darkred')
plt.title('Number of Trips per Hour in Q1')
plt.xlabel(None)
plt.ylabel(None)
plt.grid(True)
plt.xticks(range(0, 24))
plt.show()
```

This line plot shows the average number of taxi trips occurring each hour throughout Q1 of 2023.

**Observation:**

There is a noticeable peak during the early morning hours around 8 AM, which likely corresponds to the morning rush hour when people are commuting to work or school.
Another significant peak occurs around 5 PM to 6 PM, aligning with the evening rush hour.
There is a steady decline in trips after 6 PM, with the lowest activity observed during the late-night hours (2 AM - 4 AM).

**Interpretation:**

The data suggests a strong demand for taxi services during typical commuting times, with a clear drop during late-night hours when fewer people are likely traveling.

```python
months_to_plot = ['January', 'February', 'March']

fig, axs_hour = plt.subplots(len(months_to_plot), 1, figsize=(12, 10), sharex=False)
fig.suptitle('Number of Trips per Hour by Month', fontsize=16)

for i, month in enumerate(months_to_plot):
    df_month = df[df['pickup_month'] == month]
    
    hourly_trips_month = df_month.groupby('pickup_hour').size()

    axs_hour[i].plot(hourly_trips_month.index, hourly_trips_month.values, marker='o', linestyle='-', color='darkred')
    axs_hour[i].set_title(f'{month}')
    axs_hour[i].set_xlabel(None)
    axs_hour[i].set_ylabel(None)
    axs_hour[i].grid(True)
    axs_hour[i].set_xticks(range(0, 24))

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

This series of line plots represents the number of trips per hour, broken down by each month in Q1.

**January:**

Similar peaks are observed around 8 AM and 5 PM, indicating consistent commuting patterns.
The number of trips is relatively stable throughout the day, with the lowest activity during late-night hours.

**February:**

The patterns are similar to January, with noticeable peaks during commuting hours.
A slight increase in trips during the evening hours might indicate additional events or social activities.

**March:**

March shows slightly higher trip counts, especially in the evening hours. This could be due to better weather, leading to increased travel for social activities or events.

**Interpretation:**

While the general pattern remains consistent across the three months, slight variations suggest seasonal factors or specific events affecting the number of trips. March, in particular, shows a slightly higher volume, possibly indicating more favorable weather or increased activity.

<b>3. How do payment types affect total revenue and the amount of tips received?<b>

```python
df_relevant = df[['payment_type', 'total_amount', 'tip_amount']]

grouped_df = df_relevant.groupby('payment_type').sum()

grouped_df['total_revenue_percentage'] = (grouped_df['total_amount'] / grouped_df['total_amount'].sum()) * 100
grouped_df['total_tips_percentage'] = (grouped_df['tip_amount'] / grouped_df['tip_amount'].sum()) * 100

labels = ['Credit Card (1)', 'Cash (2)', 'No Charge (3)', 'Dispute (4)', 'Unknown (5)']

plt.figure(figsize=(7, 7))
plt.pie(grouped_df['total_revenue_percentage'], labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Total Revenue by Payment Type')
plt.show()
```

**Total Revenue by Payment Type**

Cash: Cash payments generated a significant portion of the total revenue, but not the highest. This indicates that while cash is still a commonly used payment method, it is not the primary source of revenue.

Credit Card: This payment type contributed the highest total revenue. The convenience of using credit cards and the potential for higher fare amounts (due to less immediate concern over spending) likely drive this result.

Other: This category, which might include alternative payment methods (e.g., vouchers, direct billing), has the lowest contribution to total revenue.

```python
plt.figure(figsize=(7, 7))
plt.pie(grouped_df['total_tips_percentage'], labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Total Tips by Payment Type')
plt.show()
```

**Total Tips by Payment Type**

Cash: Interestingly, the total tips received from cash payments are significantly lower than those from credit card payments. This might be because tipping with cash can sometimes be overlooked or rounded down more often than with card transactions.

Credit Card: Credit card payments resulted in the highest amount of tips. This is likely due to the ease of adding a tip to a card transaction and the prompts often seen on payment terminals suggesting tip percentages.

Other: The amount of tips from other payment types is negligible, possibly because these payment methods might not include tipping as a standard practice or due to the nature of the payment types.

**Interpretation**

1. Credit Card Payments:

- Total Revenue: Credit card payments account for the majority of total revenue (71.99%). This indicates a strong customer preference for using credit cards over other payment methods.

- Tips: Nearly all reported tips (99.99%) are associated with credit card transactions. The convenience of electronic tipping and possibly the prompts for tipping on card terminals significantly contribute to this outcome.

2. Cash Payments:

- Total Revenue: Cash payments contribute a smaller yet significant portion of the total revenue (27.83%). While still widely used, cash payments are less common than card payments.

- Tips: There is a negligible amount of tips associated with cash transactions. This may be due to tips not being recorded, or a lesser tendency for customers to tip when paying with cash.

2. Other Payment Types (No Charge, Dispute, Unknown):

- Total Revenue: These categories combined account for less than 1% of total revenue. They likely represent special cases, errors, or uncommon scenarios.

- Tips: Tips are practically nonexistent for these payment types, which is expected as these cases either involve no payment or are exceptions where tips are not typically collected.

**Overall Interpretation:**

The data indicates that credit card transactions not only dominate in terms of total revenue but also are overwhelmingly the primary source of tips. This suggests that customers are more likely to tip when paying by card, possibly due to the convenience and the structured prompts that encourage tipping. Conversely, cash transactions, while still a significant part of total revenue, do not significantly contribute to tip income. This differentiation highlights the impact of payment methods on overall earnings, especially in industries reliant on tips.

<b>4. What are the main factors affecting total trip revenue?</b>

```python
numerical_columns = df[['trip_distance', 'fare_amount', 'extra', 'mta_tax', 
                        'tip_amount', 'tolls_amount', 'improvement_surcharge', 'total_amount']]

correlation_matrix = numerical_columns.corr()
correlation_with_total = correlation_matrix['total_amount'].sort_values(ascending=False)
correlation_with_total

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Reds', linewidths=.5)
plt.title('Correlation Heatmap of Various Factors with Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Observations:**
1. Fare Amount (Correlation: 0.961640)

The fare amount has the strongest positive correlation with total trip revenue. This is expected since the base fare constitutes the primary component of the total charge.

2. Tip Amount (Correlation: 0.518810)

There is a moderate positive correlation between tip amount and total revenue. Tips contribute significantly to the total amount, though not as strongly as the fare itself.

3. Tolls Amount (Correlation: 0.416144)

Tolls have a moderate positive correlation with total revenue. Trips involving toll roads naturally incur additional charges, raising the total amount.

4. Extra (Correlation: 0.106340)

This likely represents additional charges (e.g., extra passengers, extra stops) and shows a weak positive correlation with total revenue.

5. Improvement Surcharge (Correlation: 0.087090)

This fixed surcharge shows a weak positive correlation with total revenue. Although it applies to most trips, its small value limits its overall impact.

6. Trip Distance (Correlation: 0.007866)

Surprisingly, trip distance has a very weak correlation with total revenue. This suggests that factors other than distance, such as fare structure and additional charges, play a more crucial role.

7. MTA Tax (Correlation: -0.056873)

There is a very slight negative correlation with the total amount, indicating that this tax does not significantly affect the total revenue.

**Interpretation:**

1. Fare Amount: As the primary component of the total charge, it naturally dictates the total revenue. Variations in fare amounts (due to different pricing structures, rates, etc.) have the most substantial impact on total revenue.

2. Tip Amount: Tips, while not mandatory, significantly increase total revenue, especially when they are substantial. The positive correlation indicates that higher tips contribute notably to the overall revenue.

3. Tolls and Extra Charges: These additional costs can significantly impact the total revenue, especially for trips involving toll roads or extra services.

4. Trip Distance: The unexpectedly low correlation between trip distance and total revenue suggests that the fare structure may not be linearly dependent on distance alone. Other factors like base fares, surcharges, and tolls play a significant role in determining the total amount.

5. Implications: Taxi service providers may focus on optimizing fare structures, encouraging tipping, and managing additional charges (like tolls and surcharges) to maximize total revenue. Understanding these factors can help in better forecasting and strategizing for revenue maximization. ​​
